#### Sorting Values of a Dataframe or Series
You can use the sort_values() function is used to sort a DataFrame or a Series based on one or more columns.
You to specify the column(s) by which you want to sort and the order (ascending or descending). Below is the syntax to use it:-

df.sort_values(col_name, axis=0, ascending=True, inplace=False, ignore_index=False)

1. col_nam - the column(s) to sort by.
2. axis - axis along which to sort. 0 for sorting by rows (default) and 1 for sorting by columns.
3. ascending - to sort in ascending order (True, default) or descending order (False).
4. inplace - to perform the sorting operation in-place (True) or return a sorted copy (False, default).
5. ignore_index - to reset the index after sorting (True) or keep the original index values (False, default).

```python
df_can.sort_values('Total', axis=0, ascending=False, inplace=False, ignore_index=False)
df_can.head()
```
